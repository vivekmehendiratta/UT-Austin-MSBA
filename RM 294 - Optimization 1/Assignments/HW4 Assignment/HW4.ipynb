```python
import numpy as np
from scipy.optimize import minimize
import pandas as pd
```

# Q1

```python
def neg_machines(lk):
    l, k = lk
    
    n_machines = 0.05 * (l**(2/3)) * (k**(1/3))
    
    return -1*n_machines
```

```python
def cons1(lk):
    l,k = lk
    
    return 100000 - 12*l - 15*k
```

```python
# now we create a dictionary telling the solver what type of constraint and what the function is
constr1 = {'type':'ineq', 'fun': cons1}
# then we put all our constraint dictionaries into a list
constraints = [constr1]
```

```python
max_machines = minimize(neg_machines,[3,2],constraints=constraints)
print(round(-max_machines.fun))
```

```python
df = pd.read_csv('homework4stocks.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
stock_mean, stock_var =  df.mean(), df.var()
```

```python
stock_cov = df.cov().values
```

```python
stock_cov
```

```python
weights = np.linspace(0,1, num = 27)
```

```python
weights @ stock_cov @ weights
```

```python
def totalVarianceQP(weights):
    return weights.T @ stock_cov @ weights
```

```python
def totalVariance(weights):
    total_var = 0
    for i in range(len(weights)):
        w_i = weights[i]
        for j in range(len(weights)):
            w_j = weights[j]
            total_var+=w_i*w_j*stock_cov.values[i][j]
    
    
    return total_var
```

```python
def cons1(weights):
    return sum(weights) - 1

def cons2(weights):
    return weights @ stock_mean - 0.01

# now we create a dictionary telling the solver what type of constraint and what the function is
constr1 = {'type':'eq', 'fun': cons1}
constr2 = {'type':'ineq', 'fun': cons2}
# then we put all our constraint dictionaries into a list
constraints = [constr1, constr2]
```

```python
weights = np.linspace(0,1, num = 27)
weights = np.ones(27)/27
best_param = minimize(totalVarianceQP, weights, constraints=constraints, bounds=[(0,1)]*27)
```

```python
best_param.x
```

```python
round(np.sqrt(best_param.fun),2)
```

```python
np.sqrt(best_param.fun)
```

## Sol : 0.08

# Solving as QP

```python
import gurobipy as gp
```

```python
threshold_return = 0.01
meanvec = stock_mean
Sigma = np.zeros((27,27))
Sigma = stock_cov
w = np.ones(27)/27

print(w @ Sigma @ w)
print(w @ meanvec)

```

```python
# constraint0: weights sum to 1
# constraint1: mean return is >= 9%

A = np.ones((2,27)) # row 0 is all 1s already

b = np.ones(2)

A[1,:] = meanvec # now change row 1
b[1] = threshold_return # now change row 1

sense = ['=','>'] 

```

```python
portMod = gp.Model()
portMod_x = portMod.addMVar(27,ub=np.array([1]*27))
portMod_con = portMod.addMConstrs(A, portMod_x, sense, b)
portMod.setMObjective(Sigma,None,0,sense=gp.GRB.MINIMIZE)

portMod.Params.OutputFlag = 0 

portMod.optimize()
```

```python
portMod_x.x
```

```python
round(np.sqrt(portMod.objVal),2)
```

# Q4

```python
df = pd.read_csv('nflratings.csv', header=None)
df.columns = ['week', 'home_team', 'away_team', 'home_team_score', 'away_team_score']
df.head()
df['score_diff'] = df['home_team_score'] - df['away_team_score']
```

#### Calculating rating using HW1 method

```python
X = (pd.get_dummies(df['home_team']).astype(int) - pd.get_dummies(df['away_team'])).values
```

```python
B = df['score_diff'].values
```

```python
M = np.matmul(X.T, X)
M[len(M) - 1:] = [1]*32
M
```

```python
p = np.matmul(X.T, B)

p[len(p) - 1] = 0
p
```

```python
M_inv = np.linalg.inv(M)
r = np.matmul(M_inv, p)
r
```

### Calculating rating using NLP

```python
def generate_df(r):
    ratings = pd.DataFrame({'team':range(1,33), 'rating':r[:-1]})
    
    df1 = df.copy()
    df1 = df1.merge(ratings, left_on='home_team', right_on='team').drop('team', axis = 1).merge(ratings, left_on='away_team', right_on='team', suffixes=('_home', '_away'))
    
    df1['home_adv'] = r[-1]
    
    df1['predited_spread'] = df1['rating_home'] - df1['rating_away'] + df1['home_adv']
    
    df1['predicted_error'] = df1['score_diff'] - df1['predited_spread']
    
    return df1
```

```python
def prediction_error(r):
    
    df2 = generate_df(r)
    
    return np.sum(df2['predicted_error']**2)
```

```python
# r = [i for i in range(70, 102)] + [4]
# prediction_error(r)

r = [i for i in range(0, 32)] + [10]
prediction_error(r)
```

```python
def cons1(r):
    return sum(r[:-1]) - 2720 

# now we create a dictionary telling the solver what type of constraint and what the function is
constr1 = {'type':'eq', 'fun': cons1}
# then we put all our constraint dictionaries into a list
constraints = [constr1]
```

```python
best_param = minimize(prediction_error,r,constraints=constraints)
```

```python
best_param.x
```

```python
prediction_error(best_param.x)
```

```python
df1 = generate_df(best_param.x)
```

```python
df1['home_win'] = df1['score_diff'] > 0
df1['predicted_home_win'] = df1['predited_spread'] > 0
```

```python
TP = sum(df1['home_win'] & df1['predicted_home_win'])
TN = sum((df1['home_win'] == False) & (df1['predicted_home_win']==False))
TP, TN
```

```python
Acc = TP+TN
Acc
```
